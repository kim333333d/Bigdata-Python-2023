##### 에어코리아 데이터 수집, 전처리

```python
import pandas as pd 
import numpy as np
```

```python
dust202210 = pd.read_excel('./raw/202210_data_past_day.xlsx')
dust202210.drop(axis= 0, index= [0], inplace=True) # 두번째 행 날리기
dust202210
```

```python
dust202210.info()
```

```python
# 컬럼명을 영문으로 
dust202210.rename(columns={'날짜':'date', 'PM10':'pm10', 'PM2.5':'pm25', 
                            '오 존':'o3', '이산화질소':'no2', '일산화탄소':'co', '아황산가스':'so2'},inplace=True)
dust202210
```

```python
dust202210 = dust202210.astype({'date':'datetime64[ns]', 'pm10':'int64', 'pm25':'int64', 'o3':'float64', 'co':'float64',
                    'no2': 'float64', 'co': 'float64', 'so2' : 'Float64'})
dust202210.tail()
```

```python
dust202210.dtypes
```

```python
# 날짜를 년월일로 분해 컬럼 생성
dust202210['year'] = dust202210['date'].dt.year
dust202210['month'] = dust202210['date'].dt.month
dust202210['day'] = dust202210['date'].dt.day
```

```python
dust202210.tail()
```

```python
# 컬럼순서 재정렬
dust202210 = dust202210[['date', 'year', 'month', 'day', 'so2', 'co', 'o3', 'no2', 'pm10', 'pm25']]
dust202210.head()
```

```python
dust202210.isnull().sum()
```

##### 기상청 데이터 확인 및 전처리

```python
weather_2210 = pd.read_excel('./raw/OBS_AWS_TIM_202210_data.xlsx')
weather_2210
```

```python
weather_2210.info()
```

```python
# 지점, 지점명 컬림 삭제
weather_2210.drop('지점', axis=1, inplace=True)
weather_2210.drop('지점명', axis=1, inplace=True)
```

```python
weather_2210.head()
```

```python
weather_2210.columns = ['date', 'temp', 'wind', 'rain', 'humid']
weather_2210.head()
```

```python
weather_2210['date'] = pd.to_datetime(weather_2210['date']).dt.date
```

```python
weather_2210 = weather_2210.astype({'date':'datetime64[ns]'})
```

```python
weather_2210.info()
```

```python
weather_2210.head()
```

```python
# 풍속 결측치를 0.0으로
weather_2210['wind'].fillna(0.0, inplace=True)
```

```python
weather_2210.info()
```

```python
# 강수량은 0.1 단위로 측정 --> 0인 데이터 0.01로 변경
weather_2210['rain'] = weather_2210['rain'].replace([0],0.01)
weather_2210['rain'].value_counts()
```

```python
weather_2210_new = weather_2210.groupby(weather_2210['date']).mean()
weather_2210_new
```

##### 미세먼지, 날씨 데이터 병합

```python
df_total = pd.merge(dust202210, weather_2210_new, how='inner', left_on='date', right_index=True)
df_total.tail()
```

```python
# 인덱스 재정리
df_total.reset_index(drop=True, inplace=True)
df_total.tail()
```
